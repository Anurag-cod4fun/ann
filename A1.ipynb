```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Activation functions

#Sigmoid Function
def sigmoid(x) -> int:
    return 1/(1 + np.exp(-x))

#relu Function
def relu(x) -> int:
    return np.maximum(0,x)

#tanh Function
def tanh(x):
    return np.tanh(x)

#softmax Function
def softmax(x):
    exp_int = np.exp(x)
    return exp_int/np.sum(exp_int)

def leaky_relu(x, alpha=0.1):
    return np.where(x >= 0, x, alpha * x)

def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))
```

```python
#generate some random data
x = np.linspace(-10, 10, 100)
```

```python
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)
y_softmax = softmax(x)
y_leaky_relu = leaky_relu(x)
y_elu = elu(x)
```

```python
plt.figure(figsize=(12,6))

plt.subplot(2,3,1)
plt.plot(x, y_sigmoid, label='Sigmoid')
plt.title("Sigmoid Activation Function")
plt.legend()
plt.subplot(2,3,2)
plt.plot(x, y_relu, label='ReLu')
plt.title("ReLu Activation Function")
plt.legend()
plt.subplot(2,3,3)
plt.plot(x, y_tanh, label='Tanh')
plt.title("Tanh Activation Function")
plt.legend()
plt.subplot(2,3,4)
plt.plot(x, y_softmax, label='Softmax')
plt.title("Softmax Activation Function")
plt.legend()
plt.subplot(2,3,5)
plt.plot(x, y_leaky_relu, label='leaky_relu')
plt.title("leaky_Relu Activation Function")
plt.legend()
plt.subplot(2,3,6)
plt.plot(x, y_elu, label='elu')
plt.title("elu Activation Function")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Activation Functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

def leaky_relu(x, alpha=0.1):
    return np.where(x >= 0, x, alpha * x)

def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

# Input range
x = np.linspace(-5, 5, 100)

# Plotting Activation Functions
plt.figure(figsize=(12, 8))
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, tanh(x), label='Tanh')
plt.plot(x, softmax(x), label='Softmax')
plt.plot(x, leaky_relu(x), label='Leaky ReLU')
plt.plot(x, elu(x), label='ELU')
plt.legend(loc='upper left')
plt.xlabel('x')
plt.ylabel('Activation')
plt.title('Activation Functions')
plt.grid(True)
plt.show()
```
